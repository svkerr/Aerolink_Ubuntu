The first column in `data` consists of badge numbers.
Create a `numpy` array where each row is a user and 365 columns are daily fetch counts (exclude badge numbers)

```python
fetches = data[:,1:366]
```

```python
fetches.shape
```

Create a numpy matrix `daily_diff` to hold the diff calculations on `fetches`.

Now for each observation (user), compute differences (`diffs`) for consecutive days and store in `daily_diff`. This is similar to differentian in Calculus. Caution - the array returned by `diff` is one element shorter than the fetches array. `diff` takes a vector as input, so we just iterate through the observations (i.e., people).

When comparing this to stock market data, each diff can be considered something like the difference between the open and closing stock price. We might want to draw upon metrics which identifies a volatile from a non-volatile stock. Volatility may be an indicator of an abnormal useage pattern.

```python
daily_diff = np.zeros((fetches.shape[0], fetches.shape[1]-1))
```

```python
for i in np.arange(fetches.shape[0]):
    daily_diff[i] = np.diff(fetches[i])
```

```python
daily_diff[360]
```

`dict_thresh`: Capture those row indices (user) whose fetch differences exceed some threshold value.

`dict_var`: Calculate variances of the differences over 365 days. Variance is a measure of how non-consistent a user fetches. 

`list_var_sorted`: Sorted list from `dict_var` from highest to lowest variances (but includes index)

`list_var_var`: unpacked variance values from `list_var_sorted`

```python
import operator
threshold = 100
dict_thresh = {}
dict_var = {}
for i in np.arange(daily_diff.shape[0]):
    L = []
    dict_var[i] = np.var(daily_diff[i])
    for j in np.arange(daily_diff.shape[1]):
        if daily_diff[i][j] > threshold:
              L.append(daily_diff[i][j])
              dict_thresh[i] = L

print max(dict_var.iteritems(), key=operator.itemgetter(1))
print
list_var_sorted = sorted(dict_var.iteritems(), key=operator.itemgetter(1),reverse=True)
print list_var_sorted[0:10]
```

```python
dict_thresh
```

Since sorting a dictionary via `iteritems()` returns a list of (index,variance) pairs, we need to unpack the variance values to plot them.

```python
%precision 3
```

```python
list_var_var= []
for i in list_var_sorted:
    x,y = i
    list_var_var.append(y)
list_var_var
```

```python
from pylab import *
rcParams['figure.figsize'] = 12,8
plt.xlabel('Individual Fetches')
plt.ylabel('Variance')
plt.plot(list_var_var[0:100])
```
